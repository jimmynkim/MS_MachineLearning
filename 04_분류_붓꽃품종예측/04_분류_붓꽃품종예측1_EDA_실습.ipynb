- 종속변수의 클래스 개수가 균등함

```python
# -----------------------------------
# 독립변수
# -----------------------------------
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

```python
# -----------------------------------
# 각 독립변수의 분포
# -----------------------------------
fig, ax = plt.subplots(2, 2, figsize=(10, 5), sharex=True, sharey=True)
axes = ax.flatten()
for i, feature in enumerate(features):
    sns.histplot(x=feature, data=df_iris, ax=axes[i])
    axes[i].set_title(feature)
plt.tight_layout()
```

```python
# -----------------------------------
# 독립변수와 종속변수간의 관계 시각화 -  boxplot
# -----------------------------------
fig, ax = plt.subplots(2, 2, figsize=(10, 10), sharex=True, sharey=True)
axes = ax.flatten()
for i, feature in enumerate(features):
    sns.boxplot(x='Species', y=feature, data=df_iris, ax=axes[i])
    axes[i].set_title(feature)
plt.tight_layout()
```

```python
# -----------------------------------
# 변수간의 관계 시각화 - pairplot
# -----------------------------------
sns.pairplot(data=df_iris, hue="Species", palette='muted');
```

- 데이터 셋의 변수 간의 관계와 데이터의 분포를 시각화하는 도구

    - 데이터셋의 모든 숫자형 변수 쌍에 대한 산점도를 표시
    - 대각선 부분에는 각 변수의 히스토그램 또는 커널밀도추정(kde) 그래프가 표시  
        해당 변수의 분포를 확인 가능
    - hue 파라미터를 사용하여 데이터셋의 특정 카테고리별로 색상을 구분하여 시각화
